```python
import pandas as pd
from pycaret.regression import setup, compare_models, blend_models, finalize_model, plot_model, predict_model, add_metric, evaluate_model
```

```python
main_train_data = pd.read_csv("HW_train_main_data.csv", parse_dates=["timestamp"])
main_test_data = pd.read_csv("HW_test_main_data.csv", parse_dates=["timestamp"])

main_train_data = main_train_data.sort_values(by=["timestamp"])
main_test_data = main_test_data.sort_values(by=["timestamp"])
```

```python
main_train_data
```

```python
setup(data=main_train_data,
          target="price",
          fold = 5)

best = compare_models(sort = 'RMSE')

pred = predict_model(best, main_test_data)
```

```python
setup(data=main_train_data.drop(['id'], axis = 1),
          target="price",
          fold = 5)

best = compare_models(sort = 'RMSE')

pred = predict_model(best, main_test_data.drop(['id'], axis = 1),)
```

```python
pred = pred.sort_values(by=["id"])
pred = pred.rename(columns={"prediction_label":"predicted_price"}).reset_index()
pred[["id","predicted_price"]].to_csv('baseline.csv', index=False)
```
